## ðŸŸ© Slicing Array (1.22.2)

```python
# Code Needed
import numpy as np
postings_list = [10, 15, 7, 20, 25, 30, 5]
postings_array = np.array(postings_list)

# Example of slicing: elements from index 1 (inclusive) to 4 (exclusive)
sliced_array = postings_array[1:4]
print(f"Original array: {postings_array}")
print(f"Sliced array (index 1 to 3): {sliced_array}")
```

## ðŸŸ¨ Highest and Lowest Salary (1.22.3)

```python
# Code Needed
salaries_list = [70000, 85000, 60000, 95000, 80000]
Highest_Salary = np.max(salaries_list)
Lowest_Salary = np.min(salaries_list)
print(f"Highest salary: {Highest_Salary}")
print(f"Lowest Salary: {Lowest_Salary}")
```
